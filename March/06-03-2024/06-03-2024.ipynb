1. You are given an array 'arr' consisting of 'n' integers which denote the position of a stall.

You are also given an integer 'k' which denotes the number of aggressive cows.

You are given the task of assigning stalls to 'k' cows such that the minimum distance between any two of them is the maximum possible.

Print the maximum possible minimum distance.

Example:
Input: 'n' = 3, 'k' = 2 and 'arr' = {1, 2, 3}

Output: 2

Explanation: The maximum possible minimum distance will be 2 when 2 cows are placed at positions {1, 3}. Here distance between cows is 2.

Sample Input 1 :
6 4
0 3 4 7 10 9


Sample Output 1 :
3


Explanation to Sample Input 1 :
The maximum possible minimum distance between any two cows will be 3 when 4 cows are placed at positions {0, 3, 7, 10}. Here distance between cows are 3, 4 and 3 respectively.


Sample Input 2 :
5 2
4 2 1 3 6


Sample Output 2 :
5


Expected time complexity:
Can you solve this in O(n * log(n)) time complexity?


Constraints :
2 <= 'n' <= 10 ^ 5
2 <= 'k' <= n
0 <= 'arr[i]' <= 10 ^ 9
Time Limit: 1 sec.

```python
#function to check if we can place all cows in the provided stalls or not
#returns boolean value
def can_we_place(stalls,distance,no_of_cows):
    count_of_cows=1      #maintaining the count of cows places in stall
    last=stalls[0]       #maintaining the position of last cow placed

    for i in range(1,len(stalls)):
        if(stalls[i]-last>=distance):
            count_of_cows+=1
            last=stalls[i]
        #if number of cows placed is more than the number of cows given then return true else false
        if count_of_cows>=no_of_cows:
            return True
    return False

#function to the actual problem of finding minimum of the maximum distance
def aggresive_cows(stalls,no_of_cows,number_of_stalls):
    stalls.sort()    #sorting the stalls position in ascending order to apply binary search
    #maintaining low high and mid as required in binary search
    low=1
    high=stalls[number_of_stalls-1]-stalls[0]
    while(low<=high):#condition to check if low doesn't cross high
        mid=(low+high)//2
        if (can_we_place(stalls, mid ,no_of_cows)==True):   #calling function can_we_place
            low=mid+1
        else:
            high=mid-1
    return high

number_of_stalls=int(input("Enter the number of stalls:"))  #user input of number of stalls
stalls=[]
for stall in range(number_of_stalls):
    stalls.append(int(input("Enter stall number:")))  #appending stalls position in stalls list

no_of_cows=int(input("Enter number of cows to place: "))
print(aggresive_cows(stalls,no_of_cows,number_of_stalls))   #printing the maximum of minimum difference between two cows
```

2.Ayush is studying for ninjatest which will be held after 'N' days, And to score good marks he has to study 'M' chapters and the ith chapter requires TIME[i] seconds to study. The day in Ayushâ€™s world has 100^100 seconds. There are some rules that are followed by Ayush while studying.

1. He reads the chapter in a sequential order, i.e. he studies i+1th chapter only after he studies ith chapter.

2. If he starts some chapter on a particular day he completes it that day itself.

3. He wants to distribute his workload over 'N' days, so he wants to minimize the maximum amount of time he studies in a day.

Your task is to find out the minimal value of the maximum amount of time for which Ayush studies in a day, in order to complete all the 'M' chapters in no more than 'N' days.

For example

if Ayush want to study 6 chapters in 3 days and the time that each chapter requires is as follows:
Chapter 1 = 30
Chapter 2 = 20
Chapter 3 = 10
Chapter 4 = 40
Chapter 5 = 5
Chapter 6 = 45

Then he will study the chapters in the following order 

| day 1 : 1 , 2 | day 2 : 3 , 4 | day 3 : 5 , 6 |
Here we can see that he study chapters in sequential order and the maximum time to study on a day is 50, which is the minimum possible in this case.
Constraints:
1 <= T <= 10
1 <= N , M <= 10 ^ 4
1 <= TIME[i] <= 10 ^ 9 
It is considered that there are infinite seconds in a day, on the planet where Ayush lives.

Time limit: 1 sec.
Sample Input 1:
1
3 5
1 2 2 3 1
Sample Output 1:
4
Explanation of sample input 1:
The ayush will read the chapter as follows,
Day 1 : 1 , 2         Time required : 3
Day 2 : 3             Time required : 2
Day 3 : 4 , 5         Time required : 4
So the maximum time in a day is 4.
Sample Input 2:
1
4 7
2 2 3 3 4 4 1 
Sample Output 2:
6
Explanation of sample input 2:
The ayush will read the chapter as follows,
Day 1 : 1 , 2          Time required : 4
Day 2 : 3 , 4          Time required : 6
Day 3 : 5              Time required : 4
Day 4 : 6 , 7          Time required : 5
So the maximum time in a day is 6.

```python
def is_possible(chapters, days, max_time):
    current_day = 1  # Initialize the current day to 1
    current_time = 0  # Initialize the current time spent on the current day to 0

    for chapter_time in chapters:
        # Check if adding the time of the current chapter to the current time
        # exceeds the maximum time allowed in a day
        if current_time + chapter_time <= max_time:
            current_time += chapter_time  # If not, add the time to the current day
        else:
            current_day += 1  # If adding the time exceeds max_time, move to the next day
            current_time = chapter_time  # Reset the current time for the new day

    return current_day <= days  # Check if the total days taken is less than or equal to the allowed days

#function to the actual problem of finding minimum of the maximum time
def minimize_max_time(chapters, days):
    low, high = 1, sum(chapters)

    while low < high:#condition to check if low doesn't cross high
        mid = (low + high) // 2
        if is_possible(chapters, days, mid):    #calling function is_possible
            high = mid
        else:
            low = mid + 1

    return low

# Input reading
size=int(input("Enter the size: "))#user input of size of time_required list
time_required=[]
for i in range(size):
    time_required.append(int(input("Enter the numbers: "))) #appending in time_required list

no_of_student=int(input("Enter number of students: "))
result = minimize_max_time(time_required, no_of_student)
print("the maximum time in a day is",result)
```

3.You have a binary tree of 'N' unique nodes and a Start node from where the tree will start to burn. Given that the Start node will always exist in the tree, your task is to print the time (in minutes) that it will take to burn the whole tree.



It is given that it takes 1 minute for the fire to travel from the burning node to its adjacent node and burn down the adjacent node.



For Example :
For the given binary tree: [1, 2, 3, -1, -1, 4, 5, -1, -1, -1, -1]
Start Node: 3

    1
   / \
  2   3
     / \
    4   5

Output: 2

Explanation :
In the zeroth minute, Node 3 will start to burn.

After one minute, Nodes (1, 4, 5) that are adjacent to 3 will burn completely.

After two minutes, the only remaining Node 2 will be burnt and there will be no nodes remaining in the binary tree. 

So, the whole tree will burn in 2 minutes.
Detailed explanation ( Input/output format, Notes, Images )
Sample Input 1 :
1 2 3 4 -1 -1 5 -1 -1 -1 -1
2    
Sample Output 1 :
3
Explanation Of Sample Input 1 :
The given binary tree will look as - 

The Start node is 2.

In the zeroth minute, Node 2 will start to burn.

After one minute, Nodes 1 and 4 that are adjacent to 2 will burn completely.

After two minutes, Node 3 that is adjacent to node 1 will burn completely.

After three minutes, the only remaining Node 5 will be burnt and there will be no nodes remaining in the binary tree i.e the whole tree will burn in 3 minutes.
Sample Input 2 :
3 1 2 5 6 -1 -1 -1 -1 -1 -1
3
Sample Output 2 :
2
Constraints :
1 <= N <= 10^5
1 <= Value of Tree Node <= 10^9
1 <= Value of Start Node <= 10^9

Time limit: 1 sec

```python
from queue import Queue      # Import the Queue library to use a queue data structure for level-order traversal

# Binary tree node class
class BinaryTreeNode:
    def __init__(self, data):
        self.data = data  # Store the node's data value
        self.left = None  # Initialize left child as None
        self.right = None  # Initialize right child as None


# Function to create parent pointers
def parent_make(parent, root, start):
    target = None
    q = [root]  # Create a queue to perform level-order traversal
    while q:
        f = q.pop(0)  # Dequeue the first element (current node)
        if f.data == start:
            target = f  # Update target if current node matches the start node
        if f.left:
            q.append(f.left)  # Add left child to the queue if it exists
            parent[f.left] = f  # Store the current node as parent of the left child
        if f.right:
            q.append(f.right)  # Add right child to the queue if it exists
            parent[f.right] = f  # Store the current node as parent of the right child
    return target


# Function to calculate time to burn the tree
def time_to_burn_tree(root, start):
    parent = {}
    target = parent_make(parent, root, start)
    q = [target]  # Initialize the queue with the target node
    burn = {target: True}  # Dictionary to track visited nodes (burned)
    cnt = 0

    while q:
        s = len(q)  # Get the size of the current queue level
        fl = 0  # Flag to check if any new nodes were added to the queue

        for i in range(s):
            temp = q.pop(0)  # Dequeue the first element (current node)

            if temp.left and not burn.get(temp.left, False):
                q.append(temp.left)  # Add left child to the queue if not burned
                burn[temp.left] = True  # Mark left child as burned
                fl = 1  # Set flag if new node was added

            if temp.right and not burn.get(temp.right, False):
                q.append(temp.right)  # Add right child to the queue if not burned
                burn[temp.right] = True  # Mark right child as burned
                fl = 1  # Set flag if new node was added

            if parent.get(temp) and not burn.get(parent[temp], False):
                q.append(parent[temp])  # Add parent to the queue if not burned
                burn[parent[temp]] = True  # Mark parent as burned
                fl = 1  # Set flag if new node was added

        if fl:
            cnt += 1  # Increment counter if new nodes were added (new level to burn)

    return cnt

# Function to take input
def take_input():
    root_data = int(input("Enter root node data (-1 for null): "))
    if root_data == -1:
        start = int(input("Enter start node data: "))
        return None, start

    root = BinaryTreeNode(root_data)  # Create the root node
    q = Queue()  # Create a queue for level-order input
    q.put(root)
    index = 1

    while not q.empty():  # Loop until the queue is empty (all nodes processed)
        current_node = q.get()  # Dequeue the first element (current node)

        left_child = int(input(f"Enter left child of {current_node.data} (-1 for null): "))
        if left_child != -1:
            left_node = BinaryTreeNode(left_child)  # Create a new node for the left child
            current_node.left = left_node  # Assign the left child to the current node
            q.put(left_node)  # Enqueue the left child for further processing

        index += 1  # Increment index (optional, for potential numbering purposes)

        right_child = int(input(f"Enter right child of {current_node.data} (-1 for null): "))
        if right_child != -1:
            right_node = BinaryTreeNode(right_child)  # Create a new node for the right child
            current_node.right = right_node  # Assign the right child to the current node
            q.put(right_node)  # Enqueue the right child for further processing

        index += 1  # Increment index (optional)

    start = int(input("Enter start node data: "))
    return root, start  # Return the constructed tree (root node) and the start node value


# Main function
def main():
    root, start = take_input()
    print("Time to burn the tree:", time_to_burn_tree(root, start))

if __name__ == "__main__":
    main()
```
