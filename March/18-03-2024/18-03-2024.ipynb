Generate CSV file with list of home price predictions.

```python
# Importing necessary libraries
import pandas as pd  # For data manipulation and analysis
import numpy as np  # For numerical computations
from sklearn.linear_model import LinearRegression  # For linear regression modeling
import matplotlib.pyplot as plt  # For data visualization

data = pd.read_csv('homeprices.csv')  # Reading 'homeprices.csv' file and storing data in DataFrame named 'data'

data  # Outputting the contents of the DataFrame 'data'

```

```python
df = pd.DataFrame(data)  # Convert loaded data to DataFrame 'df'

X = df[['area']]  # Extract input feature (X) - 'area' column   2d array
print(X)
y = df['price']   # Extract target variable (y) - 'price' column
print(y)

plt.scatter(X, y, color='blue')  # Scatter plot of 'area' vs. 'price'
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Home Prices')
plt.show()
```

```python
#Separating the features
X=df[['area']]
y=df['price']

#Initializing the model and fitting the values
model=LinearRegression()
model.fit(X,y)

#Taking in new values and predicting
new_values=[2800, 3300, 3500, 3700, 4200, 4600, 5000]
predictions=model.predict(pd.DataFrame(new_values,columns=['area']))

#Loading the predictions made in a csv file
predictions_df = pd.DataFrame({'area': new_values, 'predicted_price': predictions})
predictions_df.to_csv('PredictedValues.csv',index=False)   #saving it in a new file called PredictedValues.csv
print("Predictions have been stored in PredictedValues.csv")
```

```python
plt.scatter(new_values, predictions, color='blue')  # Scatter plot of new_values vs. predictions
plt.xlabel('Area')
plt.ylabel('Predicted price')
plt.title('Predicted Home Prices')
plt.show()  # Display the plot
```

Build a regression model and predict the per capita income for canadian citizens in year 2020.

```python
#importing necessary files
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

data = pd.read_csv('canada_per_capita_income.csv')  # Reading the CSV file into a DataFrame 'data'
data  # Displaying the contents of the DataFrame 'data'
```

```python
# Plotting the data
df = pd.DataFrame(data)  # Converting loaded data to DataFrame 'df'
X = df[['year']]
y = df['per capita income (US$)']

plt.scatter(X, y, color='blue')  # Scatter plot of 'year' vs. 'per capita income (US$)'
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada per capita income')
plt.show()
```

```python
# Initializing the model and making predictions for year 2020
model = LinearRegression()  # Creating a Linear Regression model
model.fit(X, y)  # Fitting the model to the data

year_to_predict = np.array([[2020]])  # Year for which prediction is to be made
predictions_for_2020 = model.predict(year_to_predict)  # Predicting per capita income for year 2020

# Plotting the linear regression line and the predicted value for year 2020
plt.plot(X, model.predict(X), color='red', label='Linear Regression')  # Plotting the linear regression line
plt.scatter(2020, predictions_for_2020, color='green', label='Predicted Per Capita Income for 2020')  # Plotting the predicted value for year 2020
plt.xlabel('Year')
plt.ylabel('Per Capita Income')
plt.title('Per Capita Income Prediction')
plt.legend()
plt.show()
```
