1. Sample problem of predicting home price in monroe, new jersey (USA)
Below is the table containing home prices in monroe twp, NJ. Here price depends on area (square feet), bed rooms and age of the home (in years). Given these prices we have to predict prices of new homes based on area, bed rooms and age.

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old

```python
#import necessary libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#reading the given csv file into data
data=pd.read_csv('homeprices.csv')
df=pd.DataFrame(data) #converting it to dataframe
df
```

```python
#filling the null values with median
mean_bedrooms = df['bedrooms'].mean()
df['bedrooms'].fillna(mean_bedrooms, inplace=True)
df
```

```python
# Split the data into features (X) and target variable (y)
X = df[['area', 'bedrooms', 'age']]
y = df['price']

# Split the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)
#creating object of class LinearRegression to train the model
model=LinearRegression()
model.fit(X_train,y_train)

new_values=[[3000,3,40]]
#predicting the price
predictions=model.predict(new_values)
print(f'Predicted price for the house is : {predictions[0]}')
```

Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old.

```python
second_values_to_predict=[[2500,4,5]]
#predicting the price
new_prediction=model.predict(second_values_to_predict)
print(f'Predicted price for the house is : {new_prediction[0]}')
```

2. There is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,


```python
#import necessary libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

#reading the given csv file into n_data
n_data=pd.read_csv('hiring.csv')
df1=pd.DataFrame(n_data)  #converting it to dataframe
df1
```

```python
# Calculate the median value of the 'test_score(out of 10)' column
mean_test=df1['test_score(out of 10)'].mean()

# Compute the mode (most frequent value) of the 'experience' column
mode_experience=df1['experience'].mode()
df1['test_score(out of 10)'].fillna(mean_test,inplace=True)#filling the null values with median
df1['experience'].fillna(mode_experience,inplace=True) #filling null values with mode

encoder=LabelEncoder()
# Encode the 'experience' column into numerical values
df1['experience(encoded)']=encoder.fit_transform(df1['experience'])
df1.drop(columns=['experience'], inplace=True)
df1
```

2 yr experience, 9 test score, 6 interview score

```python
# Selecting features for the independent variables (X) from the DataFrame
X=df1[['experience(encoded)','test_score(out of 10)','interview_score(out of 10)']]

# Selecting the target variable (y)
y=df1['salary($)']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.37,random_state=42)
model=LinearRegression()

# Fitting the model to the training data
model.fit(X_train,y_train)
```

```python
# Define a dictionary containing the new entry data
new_entry = {
    'experience(encoded)': [encoder.transform(['two'])[0]],  # Encode the 'experience' category 'two' and store it
    'test_score(out of 10)': [9],  # Assign a test score of 9
    'interview_score(out of 10)': [6]  # Assign an interview score of 6
}
new_data=pd.DataFrame(new_entry)
# Predict the salary using the trained model
predicted_salary=model.predict(new_data)
print(f'Predicted Salary in dollars is:{predicted_salary[0]}')
```

12 yr experience, 10 test score, 10 interview score

```python
# Define a dictionary containing the second entry data
second_entry = {
    'experience(encoded)': [encoder.fit_transform(['twelve'])[0]],  # Encode the 'experience' category 'twelve' and store it
    'test_score(out of 10)': [10],  # Assign a test score of 10
    'interview_score(out of 10)': [10]  # Assign an interview score of 10
}

second_data=pd.DataFrame(second_entry)
# Predict the salary using the trained model
second_predicted_salary=model.predict(second_data)
print(f'Predicted Salary in dollars: {second_predicted_salary[0]}')
```
