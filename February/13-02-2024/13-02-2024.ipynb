#### 8: Calculate total sale data for last year for each
product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

```python
import pandas as pd
df=pd.read_csv("company_sales_data.csv")
plt.title("Sales Data")
x=[df.facecream.sum(),df.facewash.sum(),df.toothpaste.sum(),df.bathingsoap.sum(),df.shampoo.sum(),df.moisturizer.sum()]
labels=["Facecream","Facewash","Toothpaste","Bathing Soap","Shampoo","Moisturizer"]
plt.pie(x,labels=labels, autopct='%1.1f%%')
plt.legend(loc="lower right")
plt.show()
```
