# Basic Graph

#### x = [0,1,2,3,4]
y = [0,2,4,6,8]
The line plot graph should look like this:

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,])

plt.legend()
plt.show()
```

```python
import pandas as pd
s=pd.read_csv("gas_prices.csv")
# print(s)
plt.xticks(s.Year[::3],rotation=90)
plt.plot(s.Year,s.Australia,color="red",linewidth=1,marker='o',markersize=2,label="Australia")
plt.plot(s.Year,s.Canada,color="blue",linewidth=1,marker='o',markersize=2,label="Canada")
plt.plot(s.Year,s.France,color="green",linewidth=1,marker='o',markersize=2,label="France")
plt.plot(s.Year,s["South Korea"],color="black",linewidth=1,marker='o',markersize=2,label="South Korea")
plt.plot(s.Year,s.Japan,color="yellow",linewidth=1,marker='o',markersize=2,label="Japan")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xaxis=np.array([0,1,2,3,4])
y1axis=[x*x for x in xaxis]
y2axis=[2*x for x in xaxis]

plt.plot(xaxis,y1axis)
plt.plot(xaxis,y2axis,'*:g')
plt.show()
```

```python
y=[10,20,30,40,80,90,170]
x=[2000,4000,6000,8000,12000,15000,16000]

plt.plot(x,y)

#giving title to the graph
plt.title("Population growth over the years")

#giving x and y labels
plt.xlabel("Years")
plt.ylabel("Popolation in millions")

#adding grids between the graph
plt.grid()
```

```python
y=[10,20,30,40,80,90,170]
x=[2000,4000,6000,8000,12000,15000,16000]

plt.plot(x,y)
plt.title("Population growth over the years")
plt.xlabel("Years")
plt.ylabel("Popolation in millions")

#adding only x axis grid
plt.grid(axis='x')
```

```python
x=[1,2,3,4]
y=[2,4,6,8]

plt.subplot(2,3,1)
plt.plot(x,y,"--")

x=[10,20,30,40]
y=[2,4,6,8]

plt.subplot(2,3,2)
plt.plot(x,y,":g")

x=[40,50,60,70]
y=[2,4,6,8]

plt.subplot(2,3,3)
plt.plot(x,y,"*:r")

x=[10,20,30,40]
y=[20,40,60,80]

plt.subplot(2,3,4)
plt.plot(x,y,"D:b")

x=[100,200,300,400]
y=[20,40,60,80]

plt.subplot(2,3,5)
plt.plot(x,y,"+:y")

x=[10,20,30,40]
y=[200,300,400,500]

plt.subplot(2,3,6)
plt.plot(x,y,"^:c")

plt.show()
```

```python
# Histo bars
x=["Tata","Tesla","Mahindra","Kia"]
y=[8,4,6,3]
plt.bar(x,y,width=0.5,color='green')
plt.show()
```

```python
# Histo bars
x=["Tata","Tesla","Mahindra","Kia"]
y=[8,4,6,3]

#for horizontal bars
plt.barh(x, y)
plt.show()
```

```python
x = np.random.normal(170, 10, 250)

plt.hist(x,color="red")
plt.show()
```

```python
x=[45,15,30,10]
xlabels=["Tata","Tesla","Mahindra","Kia"]
min_sale_company=[0,0,0,0.1]

plt.pie(x,labels=xlabels,explode=min_sale_company)
plt.show()
```

```python
x=[45,15,30,10]
xlabels=["Tata","Tesla","Mahindra","Kia"]

max_sale_company=[0.1,0,0,0]

plt.pie(x,labels=xlabels,explode=max_sale_company,shadow=True)
plt.legend(title="Company names")
plt.show()
```
