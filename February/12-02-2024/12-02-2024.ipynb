# Basic Graph

#### x = [0,1,2,3,4]
y = [0,2,4,6,8]
The line plot graph should look like this:

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y, 'b^--', label='2x')

x2 = np.arange(0,4.5,0.5)

plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([0,1,2,3,4,])

plt.legend()
plt.show()
```

```python
import pandas as pd
s=pd.read_csv("gas_prices.csv")
# print(s)
plt.xticks(s.Year[::3])
plt.plot(s.Year,s.Australia,color="red",linewidth=1,marker='o',markersize=2,label="Australia")
plt.plot(s.Year,s.Canada,color="blue",linewidth=1,marker='o',markersize=2,label="Canada")
plt.plot(s.Year,s.France,color="green",linewidth=1,marker='o',markersize=2,label="France")
plt.plot(s.Year,s["South Korea"],color="black",linewidth=1,marker='o',markersize=2,label="South Korea")
plt.plot(s.Year,s.Japan,color="yellow",linewidth=1,marker='o',markersize=2,label="Japan")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
xaxis=np.array([0,1,2,3,4])
y1axis=[x*x for x in xaxis]
y2axis=[2*x for x in xaxis]

plt.plot(xaxis,y1axis)
plt.plot(xaxis,y2axis,'*:g')
plt.show()
```
