# write a program to analyse student records

1. Take user input for students details and their marks
2. now analyse marks on the basis of certain thresholds that is minimum/maximum threshold 
3. find the 1st 5 topper of the batch. the first topper should be rewarded separately 
4. segregate students after calculating the avg marks in three categories (i) above maximum threshold (ii) between maximum and minimum threshold (iii) below minimum threshold 
5. the students below minimum threshold would require personal attention so now make a separate list, those who required personal attention.
6. there is also failing criteria for each subjects if any students fails in any of the subjects put them in to personal attention list

example:
Rohan has 10 out of 100 in java but have a good avg lets say 75/100 he should still be placed in personal attention list

7. Providing options for the user to choose which analysis to perform.
8. Handling edge cases, such as invalid inputs

```python
class Student:      # Class definition for Student
    def __init__(self, name):
        # Initialize a student with a name, an empty marks dictionary, and average mark set to None
        self.name = name
        self.marks = {}
        self.avg_mark = None

    def add_mark(self, subject, mark):
        # Add a mark for a specific subject to the student's marks dictionary
        self.marks[subject] = mark

    def calculate_avg_mark(self):
        # Calculate the average mark for the student based on their marks dictionary
        if self.marks:
            self.avg_mark = sum(self.marks.values()) / len(self.marks)

    def is_failing(self):
        # Check if the student is failing by examining if any mark is below 35
        for mark in self.marks.values():
            if mark < 35:
                return True
        return False

# Function to add a student record
def add_student_record():
    # Try block to handle potential errors during user input for adding a student record.
    try:
        name = input("Enter student name: ")
        student = Student(name)

        # Loop to get marks for predefined subjects and add them to the student's record.
        for subject in ["Math", "English", "Java"]:
            mark = float(input(f"Enter mark for {subject}: "))
            student.add_mark(subject, mark)

        students.append(student)
        print("Student record added successfully.")
    except ValueError:
        print("Invalid input. Please enter a valid numerical value for the mark.")


# Function to analyze student records
def analyze_student_records():
    # Try block to handle potential errors during analysis of student records.
    try:
        if not students:
            print("No student records found. Please add records first.")
            return

        above_max = []
        between_thresholds = []
        below_min = []
        personal_attention = []
        
        # Analysing and categorizing students based on maximum and minimum thresholds
        for student in students:
            student.calculate_avg_mark()

            if student.avg_mark > max_threshold:
                above_max.append(student)
            elif min_threshold <= student.avg_mark <= max_threshold:
                between_thresholds.append(student)
            else:
                below_min.append(student)

            if student.is_failing() or student.avg_mark < min_threshold:
                personal_attention.append(student)

        # Sorting students by average marks in descending order
        n = len(students)
        for i in range(n):
            for j in range(0, n-i-1):
                if students[j].avg_mark is not None and students[j+1].avg_mark is not None:
                    if students[j].avg_mark < students[j+1].avg_mark:
                        students[j], students[j+1] = students[j+1], students[j]

        # Selecting the top 5 performers
        top_performers = [student for student in students if student.avg_mark is not None][:5]   #gettng top 5 performers

        print("\nTop Performer:")
        if top_performers:
            print(f"{top_performers[0].name} (Avg: {top_performers[0].avg_mark})")       #gettng top performer

        print("\nStudents Above Maximum Threshold:")
        for student in above_max:
            print(f"- {student.name} (Avg: {student.avg_mark})")          #printing students above maximum threshold

        print("\nStudents Between Thresholds:")
        for student in between_thresholds:
            print(f"- {student.name} (Avg: {student.avg_mark})")          #printing students between maximum and minimum threshold

        print("\nStudents Below Minimum Threshold:")
        for student in below_min:
            print(f"- {student.name} (Avg: {student.avg_mark})")          #printing students below minimum threshold
        
        #List of students who need personal attention
        print("\nStudents Requiring Personal Attention:")
        for student in personal_attention:
            print(f"- {student.name} (Avg: {student.avg_mark})")
            if student.is_failing():
                failing_subjects = [subject for subject, mark in student.marks.items() if mark < 35]
                
                # printing the subject in which a particular student has failed
                if failing_subjects:
                    print(f"  Failing subjects: {', '.join(failing_subjects)}")
    except Exception as e:
        print(f"An error occurred: {e}")

        
students = []
# maximum and minimum threshold for dividing students
try:
    max_threshold = float(input("Enter the maximum threshold for student records: "))
    min_threshold = float(input("Enter the minimum threshold for student records: "))
except ValueError:
    print("Invalid input. Please enter valid numerical values for thresholds.")

while True:
    print("\nOptions:")
    print("1. Add a student record")
    print("2. Analyze student records")
    print("3. Exit")

    choice = input("Enter your choice (1-3): ")

    if choice == '1':
        add_student_record()

    elif choice == '2':
        analyze_student_records()

    elif choice == '3':
        print("Thank you for choosing us.")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 3.")
```
